```python
from sklearn.ensemble import RandomForestClassifier
from micromlgen import port
import pandas as pd

df = pd.read_excel("Sensor data esp.xlsx")

# Temperature model
X_temp = df[['Temperature']]
y_temp = df['Temp_Condition'].map({'Good': 0, 'Moderate': 1, 'Poor': 2})

model_temp = RandomForestClassifier(n_estimators=10, max_depth=5)
model_temp.fit(X_temp, y_temp)

print("Temperature model C code:")
print(port(model_temp))
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from micromlgen import port

# Load your dataset
df = pd.read_excel("Sensor data esp.xlsx")

# Drop rows with missing or invalid gas condition labels
df = df.dropna(subset=['Gas_Condition'])
valid_labels = ['Good', 'Moderate', 'Poor']
df = df[df['Gas_Condition'].isin(valid_labels)]

# Prepare features and target
X_gas = df[['Gas']]
y_gas = df['Gas_Condition'].map({'Good': 0, 'Moderate': 1, 'Poor': 2})

# Train Random Forest model
model_gas = RandomForestClassifier(n_estimators=10, max_depth=5)
model_gas.fit(X_gas, y_gas)

# Convert to C code
print("Gas model C code:")
print(port(model_gas))
```

```python
X_hum = df[['Humidity']]
y_hum = df['Humidity_Condition'].map({'Good': 0, 'Moderate': 1, 'Poor': 2})

model_hum = RandomForestClassifier(n_estimators=10, max_depth=5)
model_hum.fit(X_hum, y_hum)

print("Humidity model C code:")
print(port(model_hum))
```
